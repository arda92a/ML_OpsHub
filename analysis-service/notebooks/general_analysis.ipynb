```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("../data/Obesity_dataset.csv")
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

numeric_columns = ['Age', 'Height', 'Weight', 'NCP', 'CH2O', 'FAF', 'FCVC', 'TUE']

for col in numeric_columns:
    plt.figure(figsize=(7, 4))
    sns.histplot(df[col], kde=True, bins=30, color="skyblue", edgecolor="black", alpha=0.7)
    plt.title(f"{col} Dağılımı", fontsize=14, weight='bold')
    plt.xlabel(col, fontsize=12)
    plt.ylabel("Frekans", fontsize=12)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.grid(True, linestyle='--', alpha=0.5)
    sns.despine()
    plt.tight_layout()
    plt.show()

```

```python
categorical_columns = [
    'Gender', 'family_history_with_overweight', 'FAVC', 'CAEC',
    'SMOKE', 'SCC', 'CALC', 'MTRANS', 'NObeyesdad'
]

for col in categorical_columns:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=df, x=col, palette="Set2", order=df[col].value_counts().index, edgecolor="black")
    plt.title(f"{col} Değişkeninin Dağılımı", fontsize=14, weight='bold')
    plt.xlabel(col, fontsize=12)
    plt.ylabel("Adet", fontsize=12)
    plt.xticks(rotation=30, fontsize=10)
    plt.yticks(fontsize=10)
    plt.grid(True, axis='y', linestyle='--', alpha=0.5)
    sns.despine()
    plt.tight_layout()
    plt.show()
```

```python
continuous_vars = ['Age', 'Height', 'Weight', 'NCP', 'CH2O', 'FAF', 'FCVC', 'TUE']

for col in continuous_vars:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x='NObeyesdad', y=col, data=df, palette="Set2")
    plt.title(f"{col} Değişkeninin Obezite Seviyelerine Göre Dağılımı", fontsize=14)
    plt.xlabel("Obezite Seviyesi", fontsize=12)
    plt.ylabel(col, fontsize=12)
    plt.xticks(rotation=25)
    plt.grid(True, linestyle='--', alpha=0.4)
    sns.despine()
    plt.tight_layout()
    plt.show()
```

```python
categorical_vars = [
    'Gender', 'family_history_with_overweight', 'FAVC', 'CAEC',
    'SMOKE', 'SCC', 'CALC', 'MTRANS'
]

for col in categorical_vars:
    plt.figure(figsize=(8, 5))
    ax = sns.countplot(data=df, x=col, hue='NObeyesdad', palette="Set2")
    plt.title(f"{col} Değişkeninin Obezite Sınıflarına Göre Dağılımı", fontsize=14)
    plt.xlabel(col, fontsize=12)
    plt.ylabel("Adet", fontsize=12)
    plt.xticks(rotation=25)
    plt.grid(True, axis='y', linestyle='--', alpha=0.4)
    sns.despine()
    plt.tight_layout()
    ax.legend_.remove()  # Legend'ı kaldırıyoruz
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Legend için örnek bir grafik (tek bir plot kullanıyoruz)
fig, ax = plt.subplots(figsize=(8, 1.2))

# Boş bir countplot çiziyoruz sadece legend objesini almak için
dummy = sns.countplot(data=df, x='Gender', hue='NObeyesdad', palette="Set2", ax=ax)

# Legend objesini alıyoruz
handles, labels = ax.get_legend_handles_labels()

# Ax'ı temizliyoruz (grafik yok)
ax.clear()

# Legend'ı yeni figüre çiziyoruz
fig_leg = plt.figure(figsize=(8, 1.2))
ax_leg = fig_leg.add_subplot(111)
ax_leg.legend(handles, labels, loc='center', ncol=4, frameon=False, fontsize=12, title='Obezite Seviyesi')
ax_leg.axis('off')

# Legend figürünü kaydediyoruz
fig_leg.savefig('legend_only.png', bbox_inches='tight')

plt.close(fig)
plt.close(fig_leg)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

categorical_vars = [
    'Gender', 'family_history_with_overweight', 'FAVC', 'CAEC',
    'SMOKE', 'SCC', 'CALC', 'MTRANS'
]

# 4 satır x 2 sütun subplot düzeni
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(16, 18))
axes = axes.flatten()

# Her değişken için ayrı subplot çizimi
for i, col in enumerate(categorical_vars):
    sns.countplot(ax=axes[i], data=df, x=col, hue='NObeyesdad', palette="Set2")
    axes[i].set_title(f"{col} vs NObeyesdad", fontsize=12)
    axes[i].set_xlabel(col, fontsize=10)
    axes[i].set_ylabel("Adet", fontsize=10)
    axes[i].tick_params(axis='x', rotation=25)
    axes[i].grid(True, axis='y', linestyle='--', alpha=0.3)
    axes[i].legend_.remove()  # Legend'ı her grafikten kaldır

# Ortak legend'ı üstte ekle
handles, labels = axes[0].get_legend_handles_labels()
fig.legend(
    handles, labels,
    title="Obezite Seviyesi",
    loc='upper center',
    ncol=4,
    fontsize=10,
    title_fontsize=12,
    bbox_to_anchor=(0.5, 1)  # Buradaki 1.05 değeri boşluğu artırır
)

# Genel başlık (isteğe bağlı)

# Boşluk ayarları
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.subplots_adjust(hspace=0.5)

# Kaydetmek istersen:
# plt.savefig("categorical_vs_obezite.png", dpi=300, bbox_inches='tight')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Korelasyon matrisi için sadece sayısal değişkenleri seçiyoruz
numeric_vars = ['Age', 'Height', 'Weight', 'NCP', 'CH2O', 'FAF', 'FCVC', 'TUE']

corr = df[numeric_vars + ['NObeyesdad']].copy()

# NObeyesdad kategorik olduğundan korelasyon hesaplamak için sayısal kod verelim
# Önce hedef değişkeni sayısal hale çevirelim
df['NObeyesdad_num'] = df['NObeyesdad'].astype('category').cat.codes

corr = df[numeric_vars + ['NObeyesdad_num']].corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Sayısal Değişkenler ve Obezite Seviyesi Korelasyon Matrisi", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# 'NObeyesdad' kategorik, sayısal koda çevirelim
df['NObeyesdad_num'] = df['NObeyesdad'].astype('category').cat.codes

# Sayısal değişkenler listesi
numeric_vars = ['Age', 'Height', 'Weight', 'NCP', 'CH2O', 'FAF', 'FCVC', 'TUE']

# Korelasyonları hesapla
corr_with_target = df[numeric_vars + ['NObeyesdad_num']].corr()['NObeyesdad_num'].drop('NObeyesdad_num')

# En yüksek 3 korelasyon (mutlak değere göre)
top3 = corr_with_target.abs().sort_values(ascending=False).head(3)

print("NObeyesdad ile en yüksek korelasyona sahip 3 özellik:")
for feature in top3.index:
    corr_value = corr_with_target[feature]
    print(f"{feature}: {corr_value:.4f}")
```
